```python
# Dependencies
 import pandas as pd
```

```python
csv_path = "purchase_data.csv"
purchase_df = pd.read_csv(csv_path)

purchase_df.head()
```

```python
#Total Number of Players
sncounts = purchase_df["SN"].value_counts()
print(f"Total Players are {len(sncounts)}")
```

```python
#Purchasing Analysis (Total)
#•	Number of Unique Items
uniqueitems = purchase_df["Item ID"].value_counts()
print(f"{len(uniqueitems)} Unique Items")

#•	Average Purchase Price
purchases = purchase_df["Purchase ID"].value_counts()
purchasesnumb = len(purchases)
pricesum = purchase_df["Price"].sum()
Averageprice = pricesum/purchasesnumb
print(f"The average price of purchases is {Averageprice}")

#•	Total Number of Purchases
print(f"The total number of purchases is {purchasesnumb}")

#•	Total Revenue
print(f"The total revenue is {pricesum}")
```

```python
#Gender Demographics
groupbygender = purchase_df.groupby("Gender")

#distribution = groupbygender["Gender"].count()
distribution = groupbygender["SN"].nunique()
sumgender = pd.DataFrame({"GenderCount" : distribution})
dfgenderpct = sumgender.GenderCount / sumgender.GenderCount.sum()

sumgender = pd.DataFrame({"GenderCount" : distribution,
                          "Gender Distribution" : dfgenderpct})

sumgender

#•	Percentage and Count of Male Players
#•	Percentage and Count of Female Players
#•	Percentage and Count of Other / Non-Disclosed
```

```python
#Purchasing Analysis (Gender)
#•	The below each broken by gender 
    #o	Purchase Count
    #o	Average Purchase Price
    #o	Total Purchase Value
    #o	Average Purchase Total per Person by Gender

#groupbygender = purchase_df.groupby("Gender")

purchcount = groupbygender["Purchase ID"].count()
avgprice = groupbygender["Price"].mean()
totpurch = groupbygender["Price"].sum()
totpers = groupbygender["SN"].nunique()

sumgender2 = pd.DataFrame({"Purchases" : purchcount,
                         "Avg Price" : avgprice,
                          "Total Purchase Value " : totpurch,
                          "Avg Purchase per Person" : totpurch/totpers})

sumgender2


```

```python
#Age Demographics
#    •	The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.) 
        
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_df["Age Range"] = pd.cut(purchase_df["Age"], bins, labels=age_ranges)

groupbyagerang = purchase_df.groupby("Age Range")
agecases = groupbyagerang["SN"].nunique()
sumage = pd.DataFrame({"AgeCount" : agecases})

agepct = sumage.AgeCount / sumage.AgeCount.sum()

sumage = pd.DataFrame({"AgeCount" : agecases,
                          "Gender Distribution" : agepct})

sumage
```

```python
#        o	Purchase Count
#        o	Average Purchase Price
#        o	Total Purchase Value
#        o	Average Purchase Total per Person by Age Group

purchagecount = groupbyagerang["Purchase ID"].count()
avgageprice = groupbyagerang["Price"].mean()
totagepurch = groupbyagerang["Price"].sum()
totagepers = groupbyagerang["SN"].nunique()

sumage2 = pd.DataFrame({"Purchases" : purchagecount,
                         "Avg Price" : avgageprice,
                          "Total Purchase Value " : totagepurch,
                          "Avg Purchase per Person" : totagepurch/totagepers})

sumage2
```

```python
#Top Spenders
#    •	Identify the the top 5 spenders in the game by total purchase value, then list (in a table): 
#        o	SN
#        o	Purchase Count
#        o	Average Purchase Price
#        o	Total Purchase Value

groupbyname = purchase_df.groupby("SN")
purchcountsn = groupbyname["Purchase ID"].count()
avgpurchsn = groupbyname["Price"].mean()
totpurchsn = groupbyname["Price"].sum()



sumbyname= pd.DataFrame({"Purchase Count" : purchcountsn,
                        "Avg Purchase Price" : avgpurchsn,
                        "Total Purchase Value" : totpurchsn})

spenders = sumbyname.sort_values("Total Purchase Value", ascending=False)
spenders.head()
```

```python
#Most Popular Items
#    •	Identify the 5 most popular items by purchase count, then list (in a table): 
#        o	Item ID
#        o	Item Name
#        o	Purchase Count
#        o	Item Price
#        o	Total Purchase Value

groupbyitem = purchase_df.groupby("Item ID")
purchcountitem = groupbyitem["Purchase ID"].count()
totpurchitem = groupbyitem["Price"].sum()
itemnameunique = groupbyitem["Item Name"].unique()
Price2 = groupbyitem["Price"].unique()

sumitem = pd.DataFrame({"Item Name" : itemnameunique,
                        "Purchase Count" : purchcountitem,
                        "Item Price" : Price2,
                        "Total Purchase Value" : totpurchitem})

sumitem = sumitem.sort_values("Purchase Count", ascending=False)

sumitem.head()

```

```python
#Most Profitable Items
#    •	Identify the 5 most profitable items by total purchase value, then list (in a table): 
#        o	Item ID
#        o	Item Name
#        o	Purchase Count
#        o	Item Price
#        o	Total Purchase Value

sumitem2 = sumitem.sort_values("Total Purchase Value", ascending=False)

sumitem2.head()
```
